11. Identify and remove outliers in the km_driven column using the IQR method?

```python
# Calculate Q1, Q3, and IQR for 'km_driven'
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]

# Remove outliers
df_no_outliers = df[~((df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound))]

# Print the number of outliers and the shape of the DataFrame before and after removal
print(f"Number of outliers detected: {len(outliers)}")
print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame shape after outlier removal: {df_no_outliers.shape}")
```

13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing
year)?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Current year
current_year = 2025

# Calculate bike age
df_no_outliers['age'] = current_year - df_no_outliers['year']

# Calculate the average selling price for each age
average_selling_price_by_age = df_no_outliers.groupby('age')['selling_price'].mean().reset_index()

# Sort by age for better visualization
average_selling_price_by_age = average_selling_price_by_age.sort_values(by='age')

# Plot the average selling price by age
plt.figure(figsize=(12, 7))
sns.lineplot(x='age', y='selling_price', data=average_selling_price_by_age, marker='o')
plt.title('Average Selling Price by Bike Age')
plt.xlabel('Bike Age (Years)')
plt.ylabel('Average Selling Price')
plt.grid(True)
plt.show()

# Display the average selling price by age in a table
print("Average selling price by bike age:")
print(average_selling_price_by_age.to_markdown(index=False, numalign="left", stralign="left"))
```

14. Which bike names are priced significantly above the average price for their manufacturing year?

```python
# Calculate the average selling price for each year
average_price_by_year = df_no_outliers.groupby('year')['selling_price'].mean().reset_index()
average_price_by_year.rename(columns={'selling_price': 'average_selling_price_by_year'}, inplace=True)

# Merge the average selling price back to the main DataFrame
df_merged = pd.merge(df_no_outliers, average_price_by_year, on='year', how='left')

# Calculate the difference between actual selling price and average selling price for its year
df_merged['price_difference'] = df_merged['selling_price'] - df_merged['average_selling_price_by_year']

# Sort by price difference in descending order to find bikes significantly above average
bikes_above_average = df_merged.sort_values(by='price_difference', ascending=False)

# Display the top 10 bikes that are priced significantly above the average for their year
print("Top 10 bikes priced significantly above the average for their manufacturing year:")
print(bikes_above_average[['name', 'year', 'selling_price', 'average_selling_price_by_year', 'price_difference']].head(10).to_markdown(index=False, numalign="left", stralign="left"))
```
